```python
import pandas as pd

# Ruta al archivo CSV
ruta_del_csv = 'archive/IMDB-Movie-Data.csv'

# Cargar el conjunto de datos con pandas
dataset = pd.read_csv(ruta_del_csv)
print(dataset.columns)

# Seleccionar columnas relevantes
columnas_relevantes = ['Title', 'Year', 'Runtime (Minutes)', 'Rating', 'Votes', 'Revenue (Millions)']
dataset_seleccionado = dataset[columnas_relevantes].copy(deep=True) 

# Ahora, 'dataset_seleccionado' contiene solo las columnas relevantes del conjunto de datos
#dataset_seleccionado
dataset_seleccionado
```

Convertir estas columnas a un array de Numpy.

```python
import numpy as np

array_numpy = dataset_seleccionado.values
print(array_numpy)
```

```python
import pandas as pd
import numpy as np

print("Valores faltantes antes del reemplazo:")
print(dataset_seleccionado.isnull().sum())

valores_faltantes = dataset_seleccionado['Revenue (Millions)'].isnull().sum()
print(f"La columna 'Revenue(Millions)' tiene {valores_faltantes} valores faltantes.\n")

media_revenue = dataset_seleccionado['Revenue (Millions)'].mean()
dataset_seleccionado['Revenue (Millions)'].fillna(media_revenue, inplace=True)

#Verificar que los valores faltantes han sido reemplazados
print(dataset_seleccionado)
```
